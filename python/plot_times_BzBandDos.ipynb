```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science', 'high-vis', 'grid'])

%matplotlib inline
```

```python
number_CPUs = [1, 2, 4, 8, 16, 32, 64, 128, 240]
MPI_times = [1523.86, 791.962, 416.337, 191.86, 98.839, 49.6598, 24.7221, 14.6317,  7.0244]
OpenMP_times = [1536.28, 750.143, 384.923, 192.568, 93.199, 48.275]
```

```python
fig, axs = plt.subplots(figsize=(3.5, 2.8))
axs.scatter(number_CPUs[:6], OpenMP_times, marker="s", label="OpenMP", s=22, edgecolor="k", color="r")
axs.scatter(number_CPUs, MPI_times, marker="D", label="MPI", alpha=0.80, s=22, edgecolor="k", color="b")
axs.plot(number_CPUs,  MPI_times[0] / np.array(number_CPUs) , ls=":", lw=2, alpha=0.8, label="$\\frac{\\propto}{\\text{\#CPUs}}$", c="green")
axs.set_xscale("log")
axs.set_yscale("log")
axs.set_ylabel("time (s)")
axs.set_xlabel("\#CPUs")
axs.set_xlim(0.5, )
axs.legend()
axs.grid(which="both")

fig.tight_layout()
fig.savefig("BandsOnBZ_Computation_Time.pdf")
fig.savefig("BandsOnBZ_Computation_Time.png", dpi=600, transparent=False, facecolor="white")
```
